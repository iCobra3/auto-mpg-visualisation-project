**setting env, eyeballing the data**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('auto-mpg.csv')
df.head()
```

```python
sns.catplot(
    kind = 'bar',
    data = df,
    x = 'cylinders',
    y = 'mpg'
)
plt.show()
```

```python
sns.catplot(
    kind = 'bar',
    data = df,
    x = 'cylinders',
    y = 'mpg',
    ci = None
)
plt.show()
```

**bar chart demonstrates that cars with 4 cylinders possess highr MPG**

```python
# "the more heavy car is, the more it's mpg"
sns.relplot(
    kind = 'scatter',
    data = df,
    x = 'weight',
    y = 'mpg'
)
# sns.set_palette('Blues')
plt.show()
```

```python
#adding hue parameter to make the figure more informative 
sns.relplot(
    kind = 'scatter',
    data = df,
    x = 'weight',
    y = 'mpg',
    hue = 'model year',
    # style = 'model year'
)
plt.show()
```

```python
# customising the chart appearance
scatter = sns.relplot(
    kind = 'scatter',
    data = df,
    x = 'weight',
    y = 'mpg',
    hue = 'model year',
)

#adding title and labels
sns.set_style("whitegrid")
scatter.set(
    xlabel = "weight in KG",
    ylabel = "MPG"
)

# adding some padding to title
scatter.fig.suptitle('relation between weight to MPG' , y = 1.08)
plt.show()
```
